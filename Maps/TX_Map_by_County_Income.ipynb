```python
import plotly.figure_factory as ff
import numpy as np
import pandas as pd
import plotly.express as px
```

```python
us_county_df = pd.read_csv("https://raw.githubusercontent.com/plotly/datasets/master/minoritymajority.csv")
tx_counties_df = us_county_df[us_county_df["STNAME"] == "Texas"]
data_df = pd.read_csv("../Resources/census_medianincome_poverty.csv")
```

```python
data_df = data_df[["County", "Poverty Rate"]]
data_df["Poverty Rate"] = data_df["Poverty Rate"].round(decimals=2)
data_df = data_df.rename(columns={"County":"county"})
data_df.head()
```

```python
tx_counties_df = tx_counties_df.rename(columns={"CTYNAME":"county"})
tx_counties_df = tx_counties_df[["FIPS", "county"]]
tx_counties_df = tx_counties_df.replace(' County','',regex=True)
tx_counties_df.head()
```

```python
needed_df = pd.merge(tx_counties_df, data_df, on="county")
needed_df.head()
```

```python
values = needed_df["Poverty Rate"].tolist()
fips = needed_df["FIPS"].tolist()
endpts = list(np.mgrid[min(values):max(values):4j])
colorscale = px.colors.sequential.Reds
fig = ff.create_choropleth(
    fips=fips, values=values, scope=['Texas'], show_state_data=True,
    colorscale=colorscale, binning_endpoints=endpts, round_legend_values=True,
    plot_bgcolor='rgb(229,229,229)',
    paper_bgcolor='rgb(229,229,229)',
    legend_title='Poverty Rate Percentage by County',
    county_outline={'color': 'rgb(255,255,255)', 'width': 0.5},
    exponent_format=True,
)
fig.layout.template = None
fig.show()
```
