```python
import hvplot.pandas
import pandas as pd
import geopandas as gpd
import geoplot.crs as gcrs
import geoplot as gplt
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
school_data_df = pd.read_csv("../College Scorecard Data/tx_schools_reduced.csv")
school_data_df.head()
```

```python
school_data_df = school_data_df.dropna(subset=["Latitude"])
school_data_df.head()
```

```python
school_map = school_data_df.hvplot.points(
    "Longitude",
    "Latitude",
    geo = True,
    tiles = "CartoMidnight",
    tiles_opts = dict(alpha=0.7),
    frame_width = 700,
    frame_height = 500,
    scale = 0.01,
    color = "Type"
)
school_map
```
