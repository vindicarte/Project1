```python
import pandas as pd
from pathlib import Path
```

```python
college_data = Path("../College Scorecard Data/Most-Recent-Cohorts-Institution.csv")

college_df = pd.read_csv(college_data, low_memory=False)
college_df.head()
```

```python
college_df.columns
```

```python
pd.options.display.max_columns = None
college_df.columns
```

```python
texas_college_df = college_df.loc[college_df['STABBR'] == "TX", college_df.columns]
texas_college_df
```

```python
tx_reduced_df = texas_college_df[["INSTNM", "CITY", "STABBR", "ZIP", "CONTROL", "LATITUDE", "LONGITUDE", 
                                  "UGDS", "COSTT4_A", "COSTT4_P", "TUITIONFEE_IN", "TUITIONFEE_OUT" ]]
tx_reduced_df.head()
```

```python
tx_reduced_df = tx_reduced_df.rename(columns= {"INSTNM": "Institution Name", "CITY": "City", "STABBR": "State", 
                                               "ZIP": "Zip Code", "CONTROL": "Type", "LATITUDE": "Latitude", 
                                               "LONGITUDE": "Longitude", "UGDS": "Number of Undergraduate Students", 
                                               "COSTT4_A": "Average Cost for Academic Year Institution", 
                                               "COSTT4_P": "Average Cost for Program Year Institution", 
                                               "TUITIONFEE_IN": "Tuition Fee for In-District Students", 
                                               "TUITIONFEE_OUT": "Tuition Fee for Out-of-State Students"})
tx_reduced_df.head()
```

```python
tx_reduced_df['New Zip Code']=tx_reduced_df["Zip Code"].astype(str).str[:5]
tx_reduced_df = tx_reduced_df.reindex(columns=['Institution Name', 'City', 'State', 'Zip Code', 
                                               'New Zip Code', 'Type', 'Latitude', 'Longitude','Number of Undergraduate Students',
                                               'Average Cost for Academic Year Institution', 'Average Cost for Program Year Institution', 
                                               'Tuition Fee for In-District Students', 'Tuition Fee for Out-of-State Students'])
tx_reduced_df.head()
```

```python
tx_reduced_df.to_csv("tx_schools_reduced.csv")
```
